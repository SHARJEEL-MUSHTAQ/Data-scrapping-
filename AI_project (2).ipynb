```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Insert the URLs you want to scrape
urls = [
    'https://www.samaa.tv',
    'https://www.express.pk/',
    'https://www.express.pk/latest-news/',
    'https://www.express.pk/pakistan/',
    'https://urdu.geo.tv/',
    'https://urdu.geo.tv/latest-news',
    'https://urdu.geo.tv/category/geo-fact-check',
    'https://urdu.geo.tv/category/pakistan',
    'https://urdu.geo.tv/category/world',
    'https://urdu.geo.tv/category/sports',
    'https://urdu.geo.tv/category/entertainment',
    'https://urdu.geo.tv/category/business',
    'https://www.express.pk/blog/',
    'https://www.express.pk/business/',
    'https://www.express.pk/health/',
    'https://www.express.pk/science/',
    'https://www.express.pk/weird-news/',
    'https://www.express.pk/saqafat/',
    'https://cricketpakistan.com.pk/ur/',
    'https://www.express.pk/world/',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/%D8%A2.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/%D8%A7.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/%D8%A8.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/%D9%BE.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/ت.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/ٹ.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/ث.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/ج.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/چ.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/ح.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/خ.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/د.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/ڈ.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/ذ.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/ر.html',
    'https://www.urdupoint.com/dictionary/sort/urdu-to-english/ڑ.html'
]


    #scrape elements
documents = []
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    text = soup.get_text()
    documents.append(text)
    
    alldata=soup.text.replace('\n',"") #replace \n with space
    alldata

    RemovDigit= re.sub(r'[0-9]+', '',alldata) #Removing all Digits
    RemovDigit

    RemoveCLatters=re.sub(r'[A-Z]+', '',RemovDigit) #Removing all Capital latters
    RemoveCLatters

    RemoveSLatters=re.sub(r'[a-z]+', '',RemoveCLatters) #removing all small latters
    RemoveSLatters

    RemoveSLatters.strip()

    fstring = RemoveSLatters.replace(u'\xa0', u' ') #removing irrelivent data
    fstring

    RemoveLines=fstring.replace('—',"") 
    AfterRemoving=RemoveLines.replace("-","")
    AfterCommasRemoving=AfterRemoving.replace(",","")
    AfterCommasRemoving

    AfterPuncRemove=re.sub(r'[^\w\s]', '',AfterCommasRemoving) #punctuation removing
    AfterPuncRemove

    MyString=AfterPuncRemove.split()

    df=pd.DataFrame(MyString) #print urdu data after removing all irrelivent words
    df.head

    df.columns= ["اردو الفاظ"] #assign name to column

    df.dropna() #drop null value if exist
    df.head(1000)

    df.count() #count total words

    FinalAllWords=df.value_counts().reset_index(name='تعداد') #Count reapetation of words and assign column
    FinalAllWords
    
    print(FinalAllWords)

    #K-Mean Clustering 
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(df['اردو الفاظ'])

    k = 5  # Number of clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)

    df['Cluster'] = kmeans.labels_  # Add cluster labels to the DataFrame

    # Draw confusion matrix
    true_labels = kmeans.labels_
    cm = confusion_matrix(true_labels, kmeans.labels_)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted Cluster')
    plt.ylabel('True Cluster')
    plt.title('Confusion Matrix')
    plt.show()

    # Visualize clusters
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=X[:, 0].toarray().flatten(), y=X[:, 1].toarray().flatten(), hue=kmeans.labels_, palette='tab10')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Cluster Visualization')
    plt.legend(title='Cluster')
    plt.show()
```
