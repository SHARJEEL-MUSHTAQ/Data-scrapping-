```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt


df = pandas.read_csv(r"/Dataset.csv")
print(df)

d = {'Sunny':0,'Windy':1,'Rainy':2}
df['Weather'] = df['Weather'].map(d)
d= {'No':0,'Yes':1}
df['Parents'] = df['Parents'].map(d)
d= {'Poor':0,'Rich':1}
df['Money'] = df['Money'].map(d)
d= {'Cinema':0,'Shopping':1,'Stayin':2,'Tennis':3}
df['Decision'] = df['Decision'].map(d)

features = ['Weather','Parents','Money']
x = df[features]
y = df['Decision']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(x,y)
tree.plot_tree(dtree,feature_names=features)
```
